```python
data_obj = bnpy.data.GroupXData.read_npz('HDP-HMM input/data.npz')
```

```python
K = 10           # Number of clusters/states

# Allocation model (HDP)
gamma      =   5.0  # top-level Dirichlet concentration parameter
transAlpha =   0.5  # trans-level Dirichlet concentration parameter
startAlpha =  10.0  # starting-state Dirichlet concentration parameter
hmmKappa   =  50.0  # set sticky self-transition weight

# Observation model (1st-order Auto-regressive Gaussian)
sF = 1.0          # Set observation model prior so E[covariance] = identity
ECovMat = 'eye'
```

```python
hmmdiag_trained_model, hmmdiag_info_dict = bnpy.run(
    data_obj, 'HDPHMM', 'DiagGauss', 'memoVB',
    output_path='/tmp/mocap6/showcase-K=20-model=HDPHMM+DiagGauss-ECovMat=1*eye/',
    nLap=100, nTask=3, nBatch=1, convergeThr=0.0001,
    transAlpha=transAlpha, gamma=gamma, sF=sF, ECovMat=ECovMat,
    K=K, initname='randexamples',
    )

pickle.dump(hmmdiag_trained_model, open('hdp_hmm_model.pkl', 'wb'))
pickle.dump(hmmdiag_info_dict, open('hdp_hmm_info_dict.pkl', 'wb'))
```

```python
hmmdiag_trained_model = pickle.load(open('hdp_hmm_model.pkl', 'rb'))
hmmdiag_info_dict = pickle.load(open('hdp_hmm_info_dict.pkl', 'rb'))
```

```python
start_prob_K = hmmdiag_trained_model.allocModel.get_init_prob_vector()

# print(start_prob_K)

trans_prob_KK = hmmdiag_trained_model.allocModel.get_trans_prob_matrix()

plt.imshow(trans_prob_KK, cmap=cm.magma)
trans_prob_KK
```

```python
log_lik_seq0_TK = hmmdiag_trained_model.obsModel.calcLogSoftEvMatrix_FromPost(
    data_obj
    )

log_lik_seq0_TK.shape
```

```python
zhat_seq0_T = bnpy.allocmodel.hmm.HMMUtil.runViterbiAlg(
    log_lik_seq0_TK, np.log(start_prob_K), np.log(trans_prob_KK))

print(zhat_seq0_T.shape)
```

```python
T = np.arange(-200, 800, 2)
t = [-200,-100,0,100,200,300,400,500,600,700,800]
X0 = data_obj.X[1500:2000]
plt.figure(figsize=(10,8))
plt.subplot(311)
plt.plot(X0)
plt.xticks(np.arange(0,501,50),t)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(312)
plt.plot(T, zhat_seq0_T[1500:2000])
plt.xticks(t)
plt.xlabel('Time')
plt.ylabel('States')

plt.subplot(313)
z = zhat_seq0_T[1500:2000]
z = np.tile(z, (64,1))
cmap = cm.get_cmap('Set1', 10)
plt.imshow(z, cmap=cmap)
plt.xticks(np.arange(0,501,50),t)
plt.xlabel('Time')
plt.yticks([])
plt.colorbar()
plt.show()
```
